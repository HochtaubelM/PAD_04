Wgraj zawartość pliku orders.csv do ramki danych.
Przyjrzyj się danym, wykorzystując poznane metody (describe, info, head) zanalizuj dane.

Następnie:



1. Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
2. Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
3. Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;
4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.
5. Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.


```python
import pandas as pd

orders = pd.read_csv('orders.csv')
# 1. Zamień order_date na format daty. 
# Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień

orders['order_date'] = pd.to_datetime(orders['order_date'])
orders.head()

# 2. Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;

print('Unikalne wartości w kolumnie T-shirt category: \n', orders.tshirt_category.unique())
print('Ilość unikalnych wartości w kolumnie T-shirt category: \n', orders.tshirt_category.nunique())

# 3. Wyczyść kolumnę tshirt_category normalizując nazwy produktów. 
# Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, 
# np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;

#orders.replace({'tshirt_category':{'Hoodie':'White Hoodie','Wh Tshirt M':'White T-Shirt M', 'Bl Tshirt F':'Black T-Shirt F', 'Bl Tshirt M':'Black T-Shirt M'}}, inplace= True)

orders['tshirt_category'] = orders['tshirt_category'].apply(lambda x: x.replace('Wh Tshirt F','White T-Shirt F'))
orders['tshirt_category'] = orders['tshirt_category'].apply(lambda x: x.replace('Wh Tshirt M','White T-Shirt M'))
orders['tshirt_category'] = orders['tshirt_category'].apply(lambda x: x.replace('Bl Tshirt F','Black T-Shirt F'))
orders['tshirt_category'] = orders['tshirt_category'].apply(lambda x: x.replace('Bl Tshirt M','Black T-Shirt M'))

print('Ilość unikalnych wartości w kolumnie T-shirt category po czyszczeniu: \n', orders['tshirt_category'].nunique())

# 4. Rozdziel kolumnę tshirt category na 3 kolumny: 
# tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, 
# tshirt_size, która będzie informowała o rozmiarze koszulki, 
# tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

orders[['tshirt_colour','tshirt_type', 'tshirt_size']] = orders['tshirt_category'].str.split(expand=True)
orders.head()

# 5. Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.

orders[(orders['order_date'] >= '2014-12-31') & (orders['order_date'] <= '2016-08-02')]
```
